### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total tienda1 : ${ingreso_total_tienda1}")
print(f"Ingreso total tienda2 : ${ingreso_total_tienda2}")
print(f"Ingreso total tienda3 : ${ingreso_total_tienda3}")
print(f"Ingreso total tienda4 : ${ingreso_total_tienda4}")



import matplotlib.pyplot as plt


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
colores = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='blue')
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos ($)")
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
prod_categ_tienda1 = tienda['Categoría del Producto'].value_counts().reset_index()
prod_categ_tienda1.columns = ['Categoría del Producto Tienda 1', 'Total de Ventas']
print(prod_categ_tienda1)
prod_categ_tienda2 = tienda2['Categoría del Producto'].value_counts().reset_index()
prod_categ_tienda2.columns = ['Categoría del Producto Tienda 2', 'Total de Ventas']
print(prod_categ_tienda2)
prod_categ_tienda3 = tienda3['Categoría del Producto'].value_counts().reset_index()
prod_categ_tienda3.columns = ['Categoría del Producto Tienda 3', 'Total de Ventas']
print(prod_categ_tienda3)
prod_categ_tienda4 = tienda4['Categoría del Producto'].value_counts().reset_index()
prod_categ_tienda4.columns = ['Categoría del Producto Tienda 4', 'Total de Ventas']
print(prod_categ_tienda4)

```

# 3. Calificación promedio de la tienda


```python
prom_calif_tienda1 = tienda['Calificación'].mean()
prom_calif_tienda2 = tienda2['Calificación'].mean()
prom_calif_tienda3 = tienda3['Calificación'].mean()
prom_calif_tienda4 = tienda4['Calificación'].mean()

print(f"Calificacion Tienda 1: {prom_calif_tienda1:.2f}")
print(f"Calificacion Tienda 2: {prom_calif_tienda2:.2f}")
print(f"Calificacion Tienda 3: {prom_calif_tienda3:.2f}")
print(f"Calificacion Tienda 4: {prom_calif_tienda4:.2f}")


calificaciones = [prom_calif_tienda1, prom_calif_tienda2, prom_calif_tienda3, prom_calif_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores_tiendas = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

plt.figure(figsize=(8, 5))
bars = plt.barh(tiendas, calificaciones, color=colores_tiendas)


for bar in bars:
    plt.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}',
             va='center', ha='left', fontsize=10)


plt.title('Calificación Promedio por Tienda')
plt.xlabel('Calificación Promedio')
plt.ylabel('Tiendas')
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.show()



```

# 5. Envío promedio por tienda

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()


prom_costo_tienda1 = tienda['Costo de envío'].mean()
prom_costo_tienda2 = tienda2['Costo de envío'].mean()
prom_costo_tienda3 = tienda3['Costo de envío'].mean()
prom_costo_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envio Tienda 1: ${prom_costo_tienda1:.2f}")
print(f"Costo de envio Tienda 2: ${prom_costo_tienda2:.2f}")
print(f"Costo de envio Tienda 3: ${prom_costo_tienda3:.2f}")
print(f"Costo de envio Tienda 4: ${prom_costo_tienda4:.2f}")
```
